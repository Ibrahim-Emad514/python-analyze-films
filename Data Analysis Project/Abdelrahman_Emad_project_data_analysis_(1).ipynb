### 4.2 Missing Values
* The mean is used to replace missing values in numerical columns.
* The mode is used to replace missing values in categorical columns.

```python
# filling categorical with modedf['tagline'].fillna(df['tagline'].mode()[0], inplace=True)
df['keywords'].fillna(df['keywords'].mode()[0], inplace=True)
df['production_companies'].fillna(df['production_companies'].mode()[0], inplace=True)
df['cast'].fillna(df['cast'].mode()[0], inplace=True)
df['director'].fillna(df['director'].mode()[0], inplace=True)
df['genres'].fillna(df['genres'].mode()[0], inplace=True)
```

```python
df.drop(['id','imdb_id','homepage','revenue_adj','budget_adj'],axis=1,inplace=True)
```

##5.2 visualization and insights


Most produced?

```python
df['genres'] = df['genres'].fillna('Unknown')  # Handle missing values if necessary
genre_counts = df['genres'].str.split('|', expand=True).stack().value_counts()

# Plotting
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Each Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
# Check if the 'earn' column exists in the DataFrame. If not, create it.
if 'earn' not in df.columns:
  # Assuming 'earn' is calculated from other columns like 'budget' and 'revenue'.
  df['earn'] = df['revenue'] - df['budget']

top_10 = df.nlargest(10, 'earn')[['original_title', 'earn']].set_index('original_title')
sns.barplot(x='earn', y=top_10.index, data=top_10)
```

How the earning of movies changed over years?

```python
df.groupby('release_year')['earn'].sum().plot(kind='area',figsize=(12,8),ylabel='earning',title='Earning of movies over years')
```

How the number of produced movies changed over years?

```python
high_popularity_movies = df[df['popularity'] > 25]

# Count appearances of each actor in high popularity movies
actor_counts = {}
for cast in high_popularity_movies['cast']:
    actors = cast.split(', ')
    for actor in actors:
        actor_counts[actor.strip()] = actor_counts.get(actor.strip(), 0) + 1

# Convert to DataFrame for easier handling
actor_df = pd.DataFrame(list(actor_counts.items()), columns=['Actor', 'Count'])

# Display actors with their counts
print("Actors in High Popularity Movies:")
print(actor_df.sort_values(by='Count', ascending=False))
```

```python
df.groupby('release_year')['original_title'].count().plot(kind='area',figsize=(12,8),ylabel='Number of movies',title='Number of produced movies in each year')
```

```python
info = pd.DataFrame(df['budget'].sort_values(ascending = False))
info['original_title'] = df['original_title']
data = list(map(str,(info['original_title'])))

#extract the top 10 budget movies data from the list and dataframe.
x = list(data[:10])
y = list(info['budget'][:10])

#plot the figure and setup the title and labels.
ax = sns.pointplot(x=y,y=x)
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("Top 10 High Budget Movies",fontsize = 15)
ax.set_xlabel("Budget",fontsize = 13)
sns.set_style("darkgrid")
```
